```python
#Overfitting
#When model performs good or shows good accuracy during training and performs good during data testing as well then it is called Overfitting.

#Underfitting
#When model performs good or shows good accuracy during training and under performs or show less performance during data testing it is called Underfitting.

#Bias
#If values in features are not equal or having extreme values causing class imbalance, then model gives priorities to bigger values, it is called bias.

```
