```python
from sklearn.linear_model import LinearRegression 
```

```python
#clf for classification
#model variable 
model = LinearRegression() 
```

```python
#model ur model is empty 
```

```python
#Training Data , fit is a function that basically takes input and give output we train data with fit
model.fit(x_train,y_train) #we are taining data
```

```python
#test predict is a fuction that is usest to test the output. this will generate output from last 6 records
#this will generate predicted salaries as dataset has two columns experience and salaries
model.predict(x_test) #now we are testing
```

```python
y_test
y_pred = model.predict(x_test)
```

```python
pd.DataFrame({'y_test':y_test,'y_pred':y_pred, 'MAE':y_test-y_pred}) #show gap difference
```

```python
#use train test split 
#in the machine learning dataset is split into train set and test set
#cross validation we change chuncks of data and share to machine for analysis
#MAE actual predicted how much error (total loss)
#MSE, RMSE, R2 performs or checks overall error 
```

```python
import numpy as np 
from sklearn.metrics import mean_squared_error, r2_score
#print("Mean Absolute Error", mean_sabolute_error(y_test,y_pred))
print("Mean Sqaure Error", mean_squared_error(y_test,y_pred))
print("Root Mean Sqaure Error",np.sqrt(mean_squared_error(y_test,y_pred))) #good if result is small in num
print("R2 Socre", r2_score(y_test,y_pred)) #it is best if closer to 1
```
