#Question No. 1

a) A Pipeline combines preprocessing and modeling steps into a single object. It prevents data leakage by fitting 
preprocessing parameters only on training data and applying them to the test data. This ensures valid 
model evaluation on unseen data.

b) A Column Transformer allows you to apply different preprocessing steps to specific columns simultaneously. 
Its role is to handle mixed data types by letting you, for example, scale numerical columns while one-hot 
encoding categorical ones in a single operation.

c) Missing Value Imputation replaces null values with estimates like Mean or Median. 
Fills missing values in a dataset.

d) Train-Test Split is the process of splitting your dataset into two sets. One is used for Training (to teach the model) and 
othet is used for Testing. It is crucial because it provides an unbiased estimate of how the model will perform on real-world data.

e) EDA Stands for Exploratory Data Analysis which is the process of visualizing and summarizing data before modeling. 
Its role is to identify patterns, detect outliers, check for class imbalances like SMOTE, and understand correlations 
between features to ensure the data is clean enough for model to learn.
